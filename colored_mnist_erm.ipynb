```python
import os
import random
```

```python
import numpy as np
import torch
```

```python
from src.datasets.datasets import ColoredMNIST
from src.training.erm import train_and_test_erm
```

```python
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.random.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.random.manual_seed_all(RANDOM_SEED)
    torch.backends.cudnn.deterministic = True
```

```python
train1_set = ColoredMNIST(root=BASE_DATA_DIR, env='train1')
train1_set.plot_dataset_digits()
```
